```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
# Download user_data.csv from Google Drive into ../data/ (pip install gdown if needed)
data_dir = Path("../data")
data_dir.mkdir(exist_ok=True, parents=True)
user_data_path = data_dir / "user_data.csv"

if not user_data_path.exists():
    import gdown
    gdown.download(
        "https://drive.google.com/uc?id=1xE-GUw4zDd2_IUsj-TqdSey6e_uR-LpO",
        str(user_data_path),
        quiet=False,
    )
else:
    print(f"Using existing {user_data_path.resolve()}")
```

```python
user_data = pd.read_csv('../data/user_data.csv')
```

```python
split_date = "2026-02-01" #Fill in the date of the split YYYY-MM-DD

train = user_data[user_data["ts"] < split_date].copy()
test1 = user_data[user_data["ts"] >= split_date].copy()

print("train:", train["ts"].min(), "→", train["ts"].max(), len(train))
print("test1:", test1["ts"].min(), "→", test1["ts"].max(), len(test1))

      
```

```python
ART = Path("../artifacts")

ART.mkdir(exist_ok=True, parents=True)

train.to_parquet(ART / "train.parquet", index=False)
test1.to_parquet(ART / "test_data_pre_shift.parquet", index=False)


print("Saved to", ART.resolve())

```
